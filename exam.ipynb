```python
targets = df_train['type'].astype('category')
categorical = ['Sex']

# df_train = df_train.drop(labels='type', axis=1)

data_train_dropped = df_train.drop(labels=categorical, axis=1, inplace=False)
```

```python
# Дивимось в якій колонці є nan
columns = data_train_dropped.columns

for column in columns:
    print(column)
    print(any(data_train_dropped[column].isna()))
    print('---------------------')
```

```python
# замінимо NaN на середнє значення
for column in data_train_dropped.columns:
    data_train_dropped[column] = data_train_dropped[column].astype(np.float64)
    data_train_dropped[column].fillna(data_train_dropped[column].mean(), inplace=True)
```

```python
def plot_correlation(data, columns):
    cm = np.corrcoef(data.T)
    plt.figure(figsize=(10,10))
    heatmap = sb.heatmap(cm, cbar=True, annot=True, square=True,
                         fmt='.2f', annot_kws={'size':10}, xticklabels=columns, yticklabels=columns)
```

```python
plot_correlation(data_train_dropped, data_train_dropped.columns)
```

## Models

```python
X = data_train_dropped.values[:, :-1]
y = data_train_dropped.values[:, -1]
```

```python
# скейлимо й позбуваємось від аутлаєрів
rs = RobustScaler(with_centering=True, with_scaling=True, quantile_range=(25.0, 75.0))
X_scaled = rs.fit_transform(X)
```

```python
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import neighbors
from sklearn.linear_model import RidgeClassifier
from sklearn.svm import LinearSVC

dec_tree = tree.DecisionTreeClassifier()
forest = RandomForestClassifier()
lsvm = LinearSVC(random_state=0, max_iter=100)
knn = neighbors.KNeighborsClassifier(10)
ridge = RidgeClassifier()
models = [dec_tree, forest, lsvm, knn, ridge]
titles = ['dec_tree', 'forest', 'svm', 'knn', 'ridge']
```

```python
print('      Model                 Score')
for i in range(len(models)):
    scores = cross_val_score(models[i], X, y, cv=13)
    print(f"{titles[i]:17}:    {scores.mean():10} (+/- {scores.std() * 2})")
```

Бачимо, що найкраща модель RandomForestClassifier
